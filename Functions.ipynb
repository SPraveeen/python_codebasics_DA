<h3> function expenses </h3>

```python
expenses_sergey=[30,45,70,90]
expenses_sundar=[40,23,10,85]

#each loop instead of this use function to resue specific steps
total_sergey_expense=0
for expense in expenses_sergey:
    total_sergey_expense+=expense
print('sergey expense is',total_sergey_expense)
    
#function to do steps and we can call multiple times
def find_expenses(expenses):
    '''
    :param purose of this function we can put here, it adds expenses
    :return: total 
    '''
    total_expense=0
    for expense in expenses:
        total_expense+=expense
    return total_expense
```

```python
a=find_expenses(expenses_sergey)
b=find_expenses(expenses_sundar)

print(f'Sergey expense,{a}')
print(f'Sundar expense,{b}')

print(help(find_expenses))
```

<h3> function_arguments (*args and **kwargs) </h3>

```python
# we use *args so we can pass flexible no of values inside the input

def sum_all(*args):
    total=0
    print(args)
    for num in args:
        total+=num
    return total

total=sum_all(1,2,3,4,5)
print(total)

total1=sum_all(1,2,3,4,5,6)
print(total1)
```

```python
# we use **kwargs for key, value pair like we can add muliple values during function calling

def company_info(**kwargs):
    # if 'ticker' in kwargs:
    #     print("Ticker:" kwargs['ticker'])

    for key in kwargs:
        print(key,kwargs[key])


a=company_info(ticker='AAPL', ceo='Tim Cook', revenue='200 Billion')
b=company_info(ticker='AAPL', ceo='Tim Cook', revenue='200 Billion',pe=20)
```
