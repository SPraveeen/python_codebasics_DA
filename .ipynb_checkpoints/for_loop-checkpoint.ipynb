# expenses

```python
expenses=[1200,1300,1400,1700]
for expense in expenses:
    print(expense)
```

```python
total_expense=0
for expense in expenses:
    total_expense+=expense
print(total_expense)
```

# print month along with it use range fn

```python
expenses=[1200,1300,1400,1700]

total_expense=0
for i in range(len(expenses)):
    print(f'Month {i}, Expense : {expenses[i]} ')
    total_expense+=expenses[i]
print(total_expense)
```

```python
for i, expense in enumerate(expenses):
    print(f'Month {i}, Expense : {expense}')
```

# break statement

```python
monthly_sales=[42,38,33,38,40,45]
threshold=35
for sales_amt in monthly_sales:
    if sales_amt < threshold:
        print(sales_amt)
```

```python
# end for loop as soon as it sees first condition
for sales_amt in monthly_sales:
    if sales_amt < threshold:
        print(f'amount {sales_amt} is less that threshold')
    else:
        print(f'amount {sales_amt} is greater that threshold')
```

```python
# end for loop as soon as it sees first condition use break
for sales_amt in monthly_sales:
    if sales_amt < threshold:
        print(f'amount {sales_amt} is less that threshold')
        break
    else:
        print(f'amount {sales_amt} is greater that threshold')
```

# zip 

```python
a=[1,2,3]
b=['a','b','c']
for i in zip(a,b):
    print(i)
```

```python
tuple(zip(a,b))
```

```python
monthly_sales=[42,38,33,38,40,45]
months=['jan','feb','march','apr','may','jun']

for sales_amt,month in zip(monthly_sales,months):
    print(month,sales_amt)
```

```python
monthly_sales=[42,38,33,38,40,45]
months=['jan','feb','march','apr','may','jun']

for sales_amt,month in zip(monthly_sales,months):
    if sales_amt < threshold:
        print(f'amount {sales_amt} is less that threshold in month {month}')
        break
    else:
        print(f'amount {sales_amt} is greater that threshold in month {month}')
```

# continue

```python
for i in range(1,11):
    if i%2 !=0:
        print(i)
```

```python
for i in range(1,11):
    if i%2 ==0:
        continue
        
    print(i)
```

```python
#while loop rarely used in specific use case
n=0
while n<10:
    print(n)
    n+=1
```

```python
products=['iphone','ipad','macbook']
regions=['usa','china','india']
revenue=[20,23,45,18,17,12,12,9,5]

i=0
for product in products:
    for region in regions:
        rev=revenue[i]
        i+=1
        print(f'{product} {region} revenue: {rev}')
```
