# exercise - find average min max scores of each player in this socres.csv file

```python
#open_csv
with open('scores.csv','r') as s:
    for lines in s:
        print(lines)
```

```python
#split datas
with open('scores.csv','r') as s:
    for lines in s:
        name,match_id,scores=lines.split(',')
        print(name,match_id,scores)
```

```python
#neglect match_id and add empty dictionary
player_scores={}
with open('scores.csv','r') as s:
    for lines in s:
        name,_,scores=lines.split(',')
        player_scores[name]=[scores]
    print(player_scores)
```

```python
# append only values to taht key like for rohit many scores was there so append it
player_scores={}
with open('scores.csv','r') as s:
    for lines in s:
        name,_,scores=lines.split(',')
        if name in player_scores:
            player_scores[name].append(scores)
        else:
            player_scores[name]=[scores]
print(player_scores)
```

```python
# to remove /n and change 9 to int use
int('9\n'.strip())
```

```python
player_scores={}
with open('scores.csv','r') as s:
    for lines in s:
        name,_,scores=lines.split(',')
        scores=int(scores.strip())
        if name in player_scores:
            player_scores[name].append(scores)
        else:
            player_scores[name]=[scores]
print(player_scores)
```

```python
#find min max average
for name,runs in player_scores.items():
    min_scores=min(runs)
    max_scores=max(runs)
    avg_scores=sum(runs)/len(runs)
    print(f'{name}==>Min:{min_scores}, Max:{max_scores},Avg:{avg_scores}')
```
