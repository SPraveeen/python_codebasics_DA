- odd even

```python
n=input('enter a number')
if n%2 ==0:
    print(f'{n} is even')
else:
    print(f'{n} is odd')
```

```python
n=int(input('enter a number'))
if n%2 ==0:
    print(f'{n} is even')
else:
    print(f'{n} is odd')
```

- ternary operator

```python
n=int(input("enter a number"))
message = 'number is even' if n%2==0 else 'number is odd'
print(message)
```

```python
n=int(input('enter a number'))
if not n%2 ==0:
    print(f'{n} is odd')
else:
    print(f'{n} is even')
```
