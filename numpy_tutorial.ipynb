# lesson 3 - Matrix Operations

```python
# sales data for quarter 1
q1=np.array([
    [200,220,250], #product A
    [150,180,210], #product B
    [300,330,360] #product C
])
q2=np.array([
    [209,231,259],#product A
    [155,192,222],#product B
    [310,340,375]#product c
])
```

```python
q1+q2
```

```python
q1*q2
```

```python
q2-q1
```

```python
(q2-q1)*100/q1
```

```python
prices=np.array([
    [10,12,11],
    [8,9,10],
    [15,16,17]
])

q1_revenue=q1*prices
q1_revenue
```

```python
q1_discounts=q1_revenue*0.2
```

```python
q1_discounts
```

```python
net_revenue=q1_revenue-q1_discounts
net_revenue
```

```python
np.sum(net_revenue)
```

```python
features=np.array([
    [2000,3],
    [1800,2]
])
weights=np.array([150,50000])
np.dot(features,weights)

#2000*150 + 3*50000 = 450000
```

```python
c=np.array([1,2,3])
d=np.array([4,5,6])
np.cross(c,d)
```
